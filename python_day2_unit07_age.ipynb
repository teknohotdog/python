```python
import csv
f = open('age.csv', encoding='utf8')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해 주세요 : ')
for row in data:
    if name in row[0]:
        for i in row[3:]:
            i = i.replace(',', '')
            result.append(int(i))
print(result)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot') # 격자 무늬 스타일 지정
plt.plot(result)
plt.show() # 2번째로 실행될때 그래프 표시안됨 재실행하면 표시됨.
```
